```python
tin_chi_hoc_phan = {}
for i in data['data']['khung_chuong_trinh']:
    print(i['khoi_kien_thuc'])
    for k in i['hoc_phan_kien_thuc']:
        # if k['cac_phan_kien_thuc_chinh'] == "Ngoài khung":
        #     continue
        if k['kien_thuc_tu_chon'] is not None:
            for l in k['kien_thuc_tu_chon']:
                if l is not None:
                    tin_chi_hoc_phan[f"{l['ten_nhom_tu_chon']}"] = int(l['tin_chi_toi_thieu'])
        if k['kien_thuc_bat_buoc'] is not None:
            for m in k['kien_thuc_bat_buoc']:
                # print(m['ma_hoc_phan'])
                tin_chi_hoc_phan[m['ma_hoc_phan']] = int(float(m['so_tin_chi']))
```

```python
# xóa các các phần tử với key = TcNN2, TcNN3, TcNN4, Tc Ôn tập NN trong dict tin_chi_hoc_phan
keys_to_remove = ["TcNN2", "TcNN3", "TcNN4", "Tc Ôn tập NN"]
for key in keys_to_remove:
    if key in tin_chi_hoc_phan:
        del tin_chi_hoc_phan[key]
```

```python
so_tin_chi_theo_ky = {}
for i in data['data']['khung_chuong_trinh']:
    for k in i['hoc_phan_kien_thuc']:
        # if k['cac_phan_kien_thuc_chinh'] == "Ngoài khung":
        #     continue
        if k['kien_thuc_bat_buoc'] is not None:
            for m in k['kien_thuc_bat_buoc']:
                if m['so_tin_chi'] is not None:
                    if int(float(m['so_tin_chi'])) > 0:
                        if m['hoc_ky'] not in so_tin_chi_theo_ky:
                            so_tin_chi_theo_ky[m['hoc_ky']] = int(float(m['so_tin_chi']))
                        else:
                            so_tin_chi_theo_ky[m['hoc_ky']] += int(float(m['so_tin_chi']))
for i in data['data']['khung_chuong_trinh']:
    for k in i['hoc_phan_kien_thuc']:
        # if k['cac_phan_kien_thuc_chinh'] == "Ngoài khung":
        #     continue
        if k['kien_thuc_tu_chon'] is not None:
            for l in k['kien_thuc_tu_chon']:
                tong = 0
                tin_chi_toi_thieu = int(float(l['tin_chi_toi_thieu']))
                if l['danh_sach_mon_tu_chon'] is not None:
                    for m in l['danh_sach_mon_tu_chon']:
                        if m['so_tin_chi'] is not None:
                            if int(float(m['so_tin_chi'])) > 0:
                                tong += int(float(m['so_tin_chi']))
                                if m['hoc_ky'] not in so_tin_chi_theo_ky:
                                    so_tin_chi_theo_ky[m['hoc_ky']] = int(float(m['so_tin_chi']))
                                else:
                                    so_tin_chi_theo_ky[m['hoc_ky']] += int(float(m['so_tin_chi']))
                    if tong >= tin_chi_toi_thieu:
                        break
```

```python
import numpy as np
mon_hoc_theo_ky = {}
for i in np.arange(1, 9, 1):
    mon_hoc_theo_ky[f"ky_{i}"] = {
        'hoc_phan_bat_buoc': [],
        'hoc_phan_tu_chon': []
    }
# Thêm tất cả các môn bắt buộc vào dict, chỉ bỏ qua nếu thiếu tên hoặc học kỳ
for i in data['data']['khung_chuong_trinh']:
    for k in i['hoc_phan_kien_thuc']:
        if k['kien_thuc_bat_buoc'] is not None:
            for m in k['kien_thuc_bat_buoc']:
                ma_hp = m.get('ma_hoc_phan')
                ten_hp = m.get('ten_hoc_phan')
                hoc_ky = m.get('hoc_ky')
                so_tc = m.get('so_tin_chi')
                prerequisite = m['detail']['relations'].get('prerequisite', [])
                before = m['detail']['relations'].get('before', [])
                if ten_hp and hoc_ky:
                    mon_hoc_theo_ky[f"ky_{hoc_ky}"]['hoc_phan_bat_buoc'].append({"ma_hp": ma_hp, "ten_hp": ten_hp, "so_tc": so_tc, "hoc_ky": hoc_ky, "prerequisite": prerequisite, "before": before})
# Thêm dict nhóm tự chọn theo từng kỳ, không add từng môn riêng lẻ
for i in data['data']['khung_chuong_trinh']:
    for k in i['hoc_phan_kien_thuc']:
        if k['kien_thuc_tu_chon'] is not None:
            for l in k['kien_thuc_tu_chon']:
                nhom_tu_chon = {}
                if l['danh_sach_mon_tu_chon'] is not None:
                    # Gom các môn theo từng kỳ trong nhóm tự chọn
                    ky_dict = {}
                    for m in l['danh_sach_mon_tu_chon']:
                        ma_hp = m.get('ma_hoc_phan')
                        ten_hp = m.get('ten_hoc_phan')
                        hoc_ky = m.get('hoc_ky')
                        so_tc = m.get('so_tin_chi')
                        prerequisite = m['detail']['relations'].get('prerequisite', [])
                        before = m['detail']['relations'].get('before', [])
                        if ten_hp and hoc_ky:
                            if hoc_ky not in ky_dict:
                                ky_dict[hoc_ky] = [{"ma_hp": ma_hp, "ten_hp": ten_hp, "so_tc": so_tc, "hoc_ky": hoc_ky, "prerequisite": prerequisite, "before": before}]
                            else:
                                ky_dict[hoc_ky].append({"ma_hp": ma_hp, "ten_hp": ten_hp, "so_tc": so_tc, "hoc_ky": hoc_ky, "prerequisite": prerequisite, "before": before})
                    # Add dict nhóm tự chọn cho từng kỳ vào mon_hoc_theo_ky
                    for ky, ds_mon in ky_dict.items():
                        nhom_tu_chon = {l['ten_nhom_tu_chon']: ds_mon}
                        mon_hoc_theo_ky[f"ky_{ky}"]['hoc_phan_tu_chon'].append(nhom_tu_chon)
```

```python
# Gộp các dict có cùng key 'TcCNTT5' trong danh sách mon_hoc_theo_ky['ky_6']
def merge_dicts_with_same_key(lst, key):
    merged = []
    temp = {}
    for item in lst:
        if isinstance(item, dict) and key in item:
            if key not in temp:
                temp[key] = []
            temp[key].extend(item[key])
        else:
            merged.append(item)
    if key in temp:
        merged.append({key: temp[key]})
    return merged

for ky_hoc in mon_hoc_theo_ky.keys():
    mon_hoc_theo_ky[ky_hoc]['hoc_phan_tu_chon'] = merge_dicts_with_same_key(
        mon_hoc_theo_ky[ky_hoc]['hoc_phan_tu_chon'], 'TcCNTT5')
    mon_hoc_theo_ky[ky_hoc]['hoc_phan_tu_chon'] = merge_dicts_with_same_key(
        mon_hoc_theo_ky[ky_hoc]['hoc_phan_tu_chon'], 'TcCNTT6')
```
